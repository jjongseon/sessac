```python
URL = "https://comic.naver.com/webtoon/weekday?order=User"
response = requests.get(URL)
soup = BeautifulSoup(response.text,"html.parser")
results = soup.find_all(class_='thumb')
```

```python
results[0]
```

```python
webtoon_list = []
for result in results:
    a_tag = result.find('a')
    img_tag = result.find('img')

    href = a_tag.attrs['href']
    alt = img_tag.attrs['alt']
    webtoon_list.append({
        '제목': alt,
        '링크': href
    })
print(webtoon_list)
```

```python
webtoon_list = [{'제목': result.find('img').attrs['alt'], '링크': a_tag.attrs['href']} for result in results]
print(webtoon_list)
```
