```python
# 더보기 버튼 클릭을 위한 while문

i=0
while True:
    more_button = driver.find_element(By.ID,'more')
    more_button.click()
    if i==5:
        break
    i+=1
    

results = driver.find_elements(By.CSS_SELECTOR,'.thumb_list > li')
jungle_result = []
for result in results:
    title = result.find_element(By.CLASS_NAME, 'title').text
    category = result.find_element(By.CLASS_NAME, 'thumb_cate').text
    jungle_result.append([title,category])
print('현재 결과 개수:', len(jungle_result))
```

```python
driver.get('https://www.rocketpunch.com/jobs')
```

```python
company_list = driver.find_elements(By.CSS_SELECTOR, '.company.item')
```

```python
result = []

for company in company_list:    
    names = company.find_element(By.CLASS_NAME, 'header.name').text
    descriptions = company.find_element(By.CLASS_NAME, 'description').text

    # 채용공고 더보기 버튼 클릭
    try:
        more_button = company.find_element(By.CLASS_NAME, 'more-jobs')
        # 특정 엘레먼트로 이동
        driver.execute_script('arguments[0].scrollIntoView();', more_button )
        # 특정 엘레먼트에서 살짝 위로 이동
        driver.execute_script("window.scrollTo(0,document.body.clientHeight)-200;")
        driver.
        more_button.click()
    except:
        pass

    details = company.find_elements(By.CLASS_NAME,'job-detail')

    job_temps=[]
    for detail in details:
        title = detail.find_element(By.CSS_SELECTOR,'.job-title').text
        job_temps.append(title)

    temp={}
    temp['회사명'] = names
    temp['설명'] = descriptions
    temp ['채용공고'] = job_temps

    result.append(temp)
```

```python
page_now = 1

import time

while True:
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')

    pages = driver.find_elements(By.CSS_SELECTOR, '.pagination .computer . item')

    for page in pages:
        if page.text == str(page_now + 1):
            page_now = int(page.text)
            page.click()
            break

    time.sleep(5)

    if page_now==8:
        break
```
