실전 2 - 로켓펀치 채용 크롤링
- 회사명, 회사설명, 채용공고(회사 별 여러 개)
- 총 10페이지를 크롤링해주세요.
- https://www.rocketpunch.com/jobs 

```python
base_URL = 'https://www.rocketpunch.com/api/jobs/template?'
parameter = 'page=3'
URL = base_URL + parameter

res = requests.get(URL)
json_data = json.loads(res.text)
sp = BeautifulSoup(json_data['data']['template'],'html.parser')
```

```python
company_list = sp.select('.company.item')

len(company_list)
```

```python
results=[]
for company in company_list:
    name = company.select_one('.header.name').text.strip()
    # print(name)
    description = company.select_one('.description').text.strip()
    # print(description)
    job_details = company.select('.job-detail')

    job_temps = []


    for job_detail in job_details: 
        job_title = job_detail.select_one('.job-title').text.strip()

        job_temps.append(job_title)

    temp={}
    temp['회사명'] = name
    temp['설명'] = description
    temp['채용공고'] = job_temps
    results.append(temp)

print(results)
```

실전 3 – 삼성전자 주식 시세 크롤링
- 일별 시세 탭에 있는 데이터를 모두 크롤링해주세요.
- 성공하신분들은 총 10페이지까지 크롤링해주세요.
- https://finance.naver.com/item/sise.naver?code=005930


```python
base_URL = 'https://finance.naver.com/item/sise_day.naver?code=005930'
parameter = '&page=' +str(2)
URL = base_URL + parameter

res = requests.get(URL,headers={'User-Agent':'Mozilla/5.0'})
soup = BeautifulSoup(res.text,'html.parser')
type2_table = soup.select('.type2 tr')
```

```python
len(type2_table)
```

```python
columns = ['날짜','종가', '전일비', '시가', '고가', '저가', '거래량']
```

```python
stock_result = []
for tr in type2_table:
    try:
        date = tr.select_one('.tah').text
        nums = tr.select('.num')
        
        temp = []
        for num in nums:
            # print(num.text.strip(), end='\t')
            temp. append(num.text.strip())
        stock_result.append([date,*temp])
    except:
        pass
```
