```python
# ★★★★★ 절대경로, 상대경로 구분할 것 ★★★★★
```

```python
PDF_DIR = './실습1/pdf_files/'
file_list = os.listdir(PDF_DIR)

for i in range(len(file_list)):
    split_data = file_list[i].split('_')[1]
    original_dir = PDF_DIR + file_list[i]
    target_dir = PDF_DIR + split_data + '/'+ file_list[i]
    try:
        os.makedirs(PDF_DIR + split_data)
    except:
        pass
    shutil.move(original_dir,target_dir)
```
