```python
total_sale = 0
csv_dir = r'C:\Users\seon\Desktop\study\RPA\실습자료\실습2\csv_files'
csv_list = os.listdir(csv_dir)
for i in csv_list:
    with open(csv_dir +'/' + i, 'r') as f:
        datas = f.readlines()
        with open('orders.csv', 'a',encoding='utf-8') as f1:
            for data in datas[1:]:
                temp = data.rstrip()
                total_sale += int(temp[1])
                print(temp)
                f1.write(temp+'\n')
```
